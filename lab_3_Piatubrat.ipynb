**Виконання:**
Сортування вставленням має асимптотичну складність $O(n^2)$, оскільки для кожного елемента масиву може виконуватися до $n$ порівнянь і зсувів. Сортування бульбашкою також має складність $O(n^2)$, тому що використовує два вкладені цикли, які проходять по всьому масиву.

```python
import time
import random
import matplotlib.pyplot as plt
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
```

```python
n_values = [5, 10, 50, 100, 200, 500, 1000]
bubble_times = []
insert_times = []

for n in n_values:
    test_arr = [random.randint(0, 10000) for _ in range(n)]
    
    
    start = time.perf_counter()
    bubble_sort(test_arr.copy())
    bubble_times.append(time.perf_counter() - start)
    
   
    start = time.perf_counter()
    insertion_sort(test_arr.copy())
    insert_times.append(time.perf_counter() - start)


plt.figure(figsize=(10, 6))
plt.plot(n_values, bubble_times, label='Bubble Sort (T_bubble)', marker='o')
plt.plot(n_values, insert_times, label='Insertion Sort (T_insert)', marker='s')
plt.xlabel('Розмір масиву (n)')
plt.ylabel('Час виконання (секунди)')
plt.legend()
plt.grid(True)
plt.title('Порівняння швидкості сортування')
plt.show()
```

## Висновок
Обидва алгоритми мають складність $O(n^2)$, але на практиці сортування вставленням працює швидше за бульбашкове сортування. Зі збільшенням розміру масиву час виконання значно зростає.
