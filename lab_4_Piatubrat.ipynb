## Мета
Засвоїти технологію реалізації рекурсивних алгоритмів та оцінку їх складності.

```python
# Завдання 1: Факторіал через цикл
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(f"Факторіал 5: {factorial_iterative(5)}")
```

Асимптотична складність: $O(n)$, оскільки ми проходимо циклом від 1 до $n$ один раз.

```python
# Завдання 2: Факторіал через рекурсію
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

print(f"Факторіал 5: {factorial_recursive(5)}")
```

Асимптотична складність: Часова $O(n)$ (виконується $n$ викликів), просторова $O(n)$ (через глибину стека викликів).

```python
# Завдання 3: Фібоначчі через рекурсію
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

print(f"Число Фібоначчі (10): {fibonacci_recursive(10)}")
```

Асимптотична складність: $O(2^n)$. З кожним кроком дерево викликів розгалужується на два нових, що призводить до експоненціального зростання кількості обчислень.

```python
# Завдання 4: Сортування злиттям
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    result.extend(left[i:]); result.extend(right[j:])
    return result

test_arr = [38, 27, 43, 3, 9, 82, 10]
print(f"Відсортований масив: {merge_sort(test_arr)}")
```

Аналіз за головною теоремою:
Задача ділиться на 2 підзадачі ($a=2$), розмір кожної зменшується вдвічі ($b=2$), робота з об'єднання коштує $O(n)$. За теоремою: $T(n) = 2T(n/2) + O(n)$, що дає складність $O(n \log n)$.
