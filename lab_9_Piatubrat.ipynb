# 1. Наївний алгоритм пошуку підрядка

Наївний алгоритм посимвольно порівнює шаблон із текстом на кожній позиції.

Складність алгоритму:

O(n * m)

де:
- n — довжина тексту
- m — довжина шаблону

```python
def simple_search(substring, string):

    positions = []

    sub_len = len(substring)
    str_len = len(string)

    for start in range(str_len - sub_len + 1):

        found = True

        for offset in range(sub_len):

            if string[start + offset] != substring[offset]:
                found = False
                break

        if found:
            positions.append(start)

    return positions
```

```python
source_text = "bananabanana"
search_pattern = "ana"

matches = simple_search(search_pattern, source_text)

print("Знайдені позиції:")
print(matches)
```

# 2. Реалізація Z-функції

Z-функція показує довжину найбільшого префікса рядка, який починається з певної позиції.

```python
def build_z_array(text):

    z_values = [0 for _ in range(len(text))]

    left_border = 0
    right_border = 0

    z_values[0] = len(text)

    for current in range(1, len(text)):

        if current <= right_border:
            z_values[current] = min(
                right_border - current + 1,
                z_values[current - left_border]
            )

        while (
            current + z_values[current] < len(text)
            and text[z_values[current]] == text[current + z_values[current]]
        ):
            z_values[current] += 1

        if current + z_values[current] - 1 > right_border:
            left_border = current
            right_border = current + z_values[current] - 1

    return z_values
```

```python
sample = "mississippi"

z_result = build_z_array(sample)

print("Z-масив:")
print(z_result)
```

# 3. Пошук підрядка через Z-функцію

Алгоритм має складність:

O(n + m)

```python
def fast_substring_search(pattern, text):

    combined = pattern + "#" + text

    z_info = build_z_array(combined)

    indexes = []

    pattern_size = len(pattern)

    for pos in range(pattern_size + 1, len(combined)):

        if z_info[pos] == pattern_size:
            indexes.append(pos - pattern_size - 1)

    return indexes
```

```python
main_text = "abracadabraabra"
template = "abra"

answer = fast_substring_search(template, main_text)

print("Позиції збігів:")
print(answer)
```

# 4. Стиснення рядка за допомогою Z-функції

Ідея алгоритму:
- знайти мінімальний префікс;
- перевірити чи складається весь рядок з повторень цього префікса.

```python
def minimize_string(text):

    z_data = build_z_array(text)

    length = len(text)

    for part in range(1, length):

        if part + z_data[part] == length:

            if length % part == 0:
                return text[:part]

    return text
```

```python
example = "abcabcabcabc"

compressed = minimize_string(example)

print("Результат стиснення:")
print(compressed)
```

# Контрольні питання

## 1. Що таке префікс-функція?

Префікс-функція — це масив, який показує довжину найбільшого власного префікса, що одночасно є суфіксом.

На відміну від Z-функції, вона аналізує збіги всередині самого префікса.

## 2. Що таке Z-функція?

Z-функція — це масив, у якому кожен елемент показує довжину збігу підрядка з початком рядка.

Вона використовується для:
- пошуку підрядків;
- стиснення рядків;
- аналізу текстів.

## 3. Підходи до задачі найдовшого спільного підрядка

Основні методи:
