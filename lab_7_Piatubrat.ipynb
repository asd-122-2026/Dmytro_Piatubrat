Засвоїти основні функції та алгоритми роботи з деревами та купою засобами Python.

```python
# Реалізація бінарного дерева (списками)
def BinaryTree(r): return [r, [], []]
def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1: root.insert(1, [newBranch, t, []])
    else: root.insert(1, [newBranch, [], []])
    return root

def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1: root.insert(2, [newBranch, [], t])
    else: root.insert(2, [newBranch, [], []])
    return root

def deleteBranch(root, side):
    root[side] = []
    return root

# Приклад роботи
r = BinaryTree('A')
insertLeft(r, 'B')
insertRight(r, 'C')
print("Дерево:", r)
deleteBranch(r, 1) # Видалення лівої гілки
print("Після видалення:", r)
```

Складність: Середня/Найгірша: Search $O(n)$, Insert $O(n)$, Delete $O(n)$ (для незбалансованого дерева).

```python
heap = [0]
def parent(i): return i // 2
def swap(a, b): heap[a], heap[b] = heap[b], heap[a]

def insert(elem):
    heap.append(elem)
    index = len(heap) - 1
    while index > 1 and heap[parent(index)] < heap[index]:
        swap(parent(index), index)
        index = parent(index)

# Генерація та додавання номера (наприклад, 10)
insert(10)
print("Купа:", heap)
```

Складність: Insert $O(\log n)$, Delete $O(\log n)$, Search $O(n)$.

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, key): return hash(str(key)) % self.size

    def insert(self, key, value):
        idx = self._hash(key)
        self.table[idx].append((key, value))

    def search(self, key):
        idx = self._hash(key)
        for k, v in self.table[idx]:
            if k == key: return v
        return None

# Перевірка
ht = HashTable(10)
ht.insert(1, "Data")
print(ht.search(1))
```

Складність: Середня $O(1)$, Найгірша $O(n)$.

### Контрольні питання

**1. Чим відрізняється структура бінарне дерево від бінарного дерева пошуку?**
Бінарне дерево — це структура даних, у якій кожен вузол має не більше двох нащадків без спеціального порядку розташування.
Бінарне дерево пошуку (BST) — це дерево, яке впорядковане за правилом:
- всі елементи лівого піддерева менші за значення кореня;
- всі елементи правого піддерева більші за значення кореня.

**2. Чим відрізняється структура бінарне дерево від бінарної купи?**
Бінарне дерево — це загальна ієрархічна структура.
Бінарна купа — це спеціальне повне бінарне дерево, де для кожного вузла виконується умова:
- у max-купі батьківський елемент завжди більший за дочірні;
- у min-купі батьківський елемент завжди менший за дочірні.

**3. Які існують типи дерев? Опишіть їхні основні характеристики та переваги.**
- Бінарне дерево: кожен вузол має до двох нащадків.
- Бінарне дерево пошуку (BST): дозволяє ефективно шукати дані.
- AVL-дерево та червоно-чорне дерево: самобалансовані структури, що підтримують висоту $O(\log n)$.
- B-дерево: використовується в базах даних для роботи з великими обсягами на дисках.
- Переваги: ієрархічне зберігання та швидка навігація по даних.

- Організація файлових систем (директорії та файли).
- Обробка HTML/XML-документів (DOM-дерево).
- Побудова таблиць маршрутизації в мережах.
- Пошук та сортування даних у базах даних.

**5. Як організована купа? Опишіть алгоритми додавання та вилучення елементів з купи.**
Купа організована як масив, що представляє повне бінарне дерево.
- Додавання: новий елемент додається в кінець масиву, потім порівнюється з батьком і "спливає" вгору (swap), поки властивість купи не відновиться.
- Вилучення: видаляється корінь, на його місце ставиться останній елемент масиву, який потім "потопає" вниз (maxHeapify/sift-down), доки не займе правильну позицію.

**6. Які задачі можна ефективно вирішити за допомогою купи? Наведіть приклади.**
- Реалізація пріоритетних черг (де найважливіші задачі виконуються першими).
- Алгоритм Дейкстри для знаходження найкоротшого шляху на графі.
- Сортування Heap Sort (ефективне сортування без виділення додаткової пам'яті).

**7. Як геш-функція використовується для зберігання та пошуку даних в хеш-таблиці?**
Геш-функція перетворює довільний ключ (рядок, число тощо) на ціле число — індекс у масиві.
- Зберігання: за допомогою функції обчислюється індекс, куди записується значення.
- Пошук: за тим самим ключем знову обчислюється індекс, що дає миттєвий доступ ($O(1)$) до потрібної комірки.

**8. Опишіть методи вирішення колізій в хеш-таблицях. Які їхні переваги та недоліки?**
- Ланцюжкове гешування: у кожній комірці таблиці зберігається список елементів.
    - Переваги: простота реалізації.
    - Недоліки: вимагає додаткової пам'яті під списки.
- Відкрита адресація: при колізії шукається наступна вільна комірка.
    - Переваги: економія пам'яті (немає списків).
    - Недоліки: складніше видалення елементів та ризик "кластеризації".
