## Мета
Засвоїти головні функції та алгоритми роботи зі стеком і чергою засобами Python, навчитися оцінювати їх асимптотичну складність.

```python
class Stack:
    def __init__(self): self.items = []
    def isEmpty(self): return self.items == []
    def push(self, item): self.items.append(item)
    def pop(self): return self.items.pop()
    def peek(self): return self.items[-1]
    def size(self): return len(self.items)

# Завдання: функція pop_n
def pop_n(stack, n):
    for i in range(min(n + 1, stack.size())):
        print("Видалено:", stack.pop())

# Перевірка
s = Stack()
for i in [1, 2, 3, 4, 5]: s.push(i)
print("Початковий стек:", s.items)
pop_n(s, 2)
print("Стек після видалення:", s.items)
```

```python
class Queue:
    def __init__(self): self.items = []
    def isEmpty(self): return self.items == []
    def enqueue(self, item): self.items.insert(0, item)
    def dequeue(self): return self.items.pop()
    def size(self): return len(self.items)

# Завдання: функція print_n
def print_n(q, n):
    print("Елементи черги (від початку до n):")
    for i in range(min(n + 1, q.size())):
        print(q.items[-(i + 1)])

# Перевірка
q = Queue()
for i in [10, 20, 30, 40]: q.enqueue(i)
print_n(q, 2)
```

## Аналіз асимптотичної складності

| Операція | Стек (середній/найгірший) | Черга (середній/найгірший) |
| :--- | :--- | :--- |
| **Insert** | $O(1)$ | $O(n)$ |
| **Delete** | $O(1)$ | $O(1)$ |
| **Search** | $O(n)$ | $O(n)$ |

*Примітка: Черга реалізована через `insert(0, item)`, що вимагає зміщення елементів ($O(n)$).*

### Контрольні питання

**1. Що таке стек і які операції можна виконувати зі стеком?**
Стек — це лінійна структура даних, що працює за принципом LIFO (Last In, First Out — останній прийшов, перший пішов).
Основні операції:
- push: додавання елемента у вершину стека;
- pop: видалення та повернення елемента з вершини;
- peek: перегляд верхнього елемента без його видалення;
- isEmpty: перевірка, чи порожній стек;
- size: повернення кількості елементів у стеку.

**2. Яка основна відмінність між стеком та чергою?**
Основна відмінність полягає в порядку обробки елементів:
- Стек працює за принципом LIFO (Last In, First Out).
- Черга працює за принципом FIFO (First In, First Out — перший прийшов, перший пішов).
У стеку додавання та видалення відбуваються з одного кінця (вершини), у черзі — з різних (додавання в кінець, видалення з початку).

**3. Як реалізувати стек за допомогою масиву і зв’язаного списку? Переваги та недоліки.**
- Реалізація через масив: елементи зберігаються послідовно. 
    - Переваги: простота реалізації, швидкий доступ за індексом. 
    - Недоліки: фіксований розмір, необхідність виділення пам'яті заздалегідь.
- Реалізація через зв'язаний список: кожен елемент зберігає посилання на наступний. 
    - Переваги: динамічний розмір, ефективна пам'ять. 
    - Недоліки: складніша структура, додаткова пам'ять на посилання.

**4. Які є застосування стека та черги в програмуванні і реальному житті?**
- Застосування стека: механізм Undo в редакторах, стек викликів функцій у мовах програмування, перевірка збалансованості дужок, історія навігації в браузері.
- Застосування черги: черги на друк документів, обробка запитів до сервера (планувальник задач), системи обслуговування клієнтів (черги в банках або магазинах).
